## 主要结论前置

1. [复购率高达 89.20%，客户粘性很高。](#交易指标)
2. [用户的活跃度和销售额在黑五期间（11.21~11.24）达季度最高峰，而在圣诞节（11.25）后持续走低。](#按日期分析)
3. [活动期间，除了流量、销量，用户客单价也会明显提高，可适当推高价高利润商品。](#按日期分析)
4. [11 时和 19 时左右是每日用户活跃与成交的高峰期，客单价也高于其它时段，但各环节转化率没有提升，可加强该时段的运营（短时活动、后台推送）。](#按时段分析)
5. [用户浏览与重复浏览的转化率低，商品的首次曝光和重复曝光策略应该进行调整。](#用户行为路径分布)
6. [用户日常粘性有待提高，以期提高单个用户的 LTV。](#用户活跃情况)
7. [用户加购后购买的转化率明显低于没有后续操作和取消购物车，推测用户可能以“加购”代替“商品收藏”或“价格观察”，不一定代表高购买意愿。](#用户行为路径分布)
8. [虽然客户复购率高，但客户首日购买的商品数量明显高于复购日购买的商品数量，推测是大量购入试用再选品复购，因此做好首单服务和分析很重要。](#用户购买情况)
9. [客户购买后的间隔天数越小，复购的概率越高。](#复购间隔天数分布)
10. [排除均价的因素，品牌流量的高低几乎决定销量的高低。近 80% 的品牌处于低流量-低销量的状况，可从提高流量的角度入手改善。](#品牌矩阵分析)

```python
import numpy as np
import pandas as pd
pd.set_option('display.float_format',lambda x : '%.6f' % x)
pd.set_option("max_rows", 100)

from pyecharts.charts import *
from pyecharts.components import Table
from pyecharts.commons.utils import JsCode
import pyecharts.options as opts
from pyecharts.globals import ThemeType

import os
import gc
import datetime


def show_time_trend(uv_df, behaviour_df, bar_title='UV', bar_subtitle='-', line_title='Behaviour', line_subtitle='-', datazoom_opts_show=False, is_show=True):
    """展示PV/UV柱状图和行为指标折线图"""
    uv_bar = (
        Bar()
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=bar_title,
                subtitle=bar_subtitle,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
        )
        .add_xaxis(uv_df.index.astype('str').tolist())
    )
    for c in uv_df.columns:
        uv_bar.add_yaxis(c, uv_df[c].tolist())
        uv_bar.set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            ),
            markline_opts=opts.MarkLineOpts(
                data=[
                    opts.MarkLineItem(name=f"{c}均值", type_="average")
                ],
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
                linestyle_opts=opts.LineStyleOpts(
                    opacity=0.5,
                    type_='dashed'
                ),
            ),
        )
    if datazoom_opts_show:
        uv_bar.set_global_opts(datazoom_opts=opts.DataZoomOpts(is_show=datazoom_opts_show))
    
    behaviour_line = (
        Line()
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=line_title,
                subtitle=line_subtitle,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
        )
        .add_xaxis(behaviour_df.index.astype(str).tolist())
    )
    for b in behaviour_df.columns:
        behaviour_line.add_yaxis(b, behaviour_df[b].tolist())
        behaviour_line.set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=2),
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            ),
            markline_opts=opts.MarkLineOpts(
                data=[
                    opts.MarkLineItem(name=f"{b}均值", type_="average")
                ],
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
                linestyle_opts=opts.LineStyleOpts(
                    opacity=0.5,
                    type_='dashed'
                ),
            ),
        )

    page = (
        Page()
        .add(uv_bar, behaviour_line)
    )
    return page.render_notebook() if is_show else page

def double_y(data, left, right, title='Double y-axis', left_kind=Line(), right_kind=Line(), is_show=True):
    """双 y 轴"""
    left_chart = (
        Line()
        .add_xaxis(data.index.astype('str').tolist())
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=title,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                name=f"{data.index.name}",
                name_location='middle',
                name_gap=30,
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
            yaxis_opts=opts.AxisOpts(
                name=f"{'/'.join(left)}",
            ),
        )
        .extend_axis(
            yaxis=opts.AxisOpts(
                name=f"{'/'.join(right)}",
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            )
        )
    )

    for l in left:
        left_chart.add_yaxis(
            l,
            data[l].tolist(),
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=2),
        )

    right_chart = (
        Line()
        .add_xaxis(data.index.astype('str').tolist())
    )

    for r in right:
        right_chart.add_yaxis(
            r,
            data[r].tolist(),
            yaxis_index=1,
            linestyle_opts=opts.LineStyleOpts(curve=360, type_='dotted'),
            label_opts=opts.LabelOpts(is_show=False),
        )

    left_chart.overlap(right_chart)
    return left_chart.render_notebook() if is_show else left_chart

def draw_pareto(data, tittle='帕累托图', x_name='', is_show=True):
    """帕累托图"""
    data[f'{data.columns[0]}-累计占比'] = np.round(data.values[:, 0].cumsum() / data.values[:, 0].sum(), 2)
    
    bar = (
        Bar()
        .set_global_opts(
            title_opts=opts.TitleOpts(title=tittle),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                name=x_name,
                name_location='center',
                name_gap=30,
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
            yaxis_opts=opts.AxisOpts(
                name=data.columns[0],
                type_="value",
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
                axistick_opts=opts.AxisTickOpts(is_show=True),
            ),
            legend_opts=opts.LegendOpts(
                type_="scroll",
            ),
        )
        .extend_axis(
            yaxis=opts.AxisOpts(
                name=data.columns[1],
                type_="value",
                min_=0,
                max_=1.1,
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            )
        )
        .add_xaxis(data.index.astype(str).tolist())
        .add_yaxis(
            data.columns[0], 
            data.iloc[:, 0].tolist(),
            z=0
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )

    cumsum_line = (
        Line()
        .add_xaxis(data.index.astype(str).tolist())
        .add_yaxis(
            data.columns[1], 
            data.iloc[:, 1].tolist(),
            yaxis_index=1,
            linestyle_opts=opts.LineStyleOpts(width=1),
            markline_opts=opts.MarkLineOpts(
                data=[opts.MarkLineItem(y=0.8)],
                linestyle_opts=opts.LineStyleOpts(
                    width=1,
                    type_='dashed',
                    opacity=0.8,
                ),
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
            ),
            z=1
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )
    
    return bar.overlap(cumsum_line).render_notebook() if is_show else bar.overlap(cumsum_line)

def draw_behaviour_funnel(data, behaviour_lst, overall=True, is_show=True, title='Funnel'):
    behaviour_df = pd.DataFrame(data[behaviour_lst].sum(), index=behaviour_lst, columns=['n'])
    if overall:
        behaviour_df['转化率'] = behaviour_df['n'] / behaviour_df.iloc[0, 0]
    else:
        behaviour_df['转化率'] = behaviour_df['n'].shift(-1) / behaviour_df['n']
        behaviour_df['转化率'] = behaviour_df['转化率'].shift(1)
        behaviour_df.iloc[0, 1] = 1
    behaviour_df['转化率'] = behaviour_df['转化率'].apply(lambda x: format(x, ".2%"))
    behaviour_df['转化环节'] = behaviour_df.index + ': ' + behaviour_df['转化率']

    behaviour_funnel = (
        Funnel()
        .add(
            '',
            behaviour_df[['转化环节', 'n']].values,
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=title,
                subtitle='转化率计算方式：每个环节以第一个环节作为计算基准' if overall else '计算方式：每个环节以前一个环节作为计算基准',
            ),
            legend_opts=opts.LegendOpts(
                pos_left='right',
                pos_top='bottom',
                orient='vertical'
            ),
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(
                position='inner',
                formatter='{b}',
            )
        )
    )
    return behaviour_funnel.render_notebook() if is_show else behaviour_funnel

# title = '浏览-加购-购买整体转化'
# draw_behaviour_funnel(data, ['view', 'cart', 'purchase'], title=title)
# title = '浏览-加购-移出购物车整体转化'
# draw_behaviour_funnel(data, ['view', 'cart', 'remove_from_cart'], title=title)


DATA_PATH = '../data/'
CLEAN_PATH = '../clean/'
CACHE_PATH = '../cache/'
DATA_FILES = ['2019-Oct.csv', '2019-Nov.csv', '2019-Dec.csv',]
cleaned_target = f"{CLEAN_PATH}processed_data.h5"

data = pd.read_hdf(cleaned_target, key='q4')
data = pd.concat([data, pd.get_dummies(data['event_type'])], axis=1)
data['turnovers'] = data['price'] * data['purchase']
data['date'] = data['event_time'].dt.date
data['month'] = data['event_time'].dt.month
data['week'] = data['event_time'].dt.isocalendar().week
data['weekday'] = data['event_time'].dt.weekday
weekday_mapper = ['星期一', '星期二', '星期三', '星期四', '星期五', '星期六', '星期日', ]
weekday_mapper = dict(zip([i for i in range(len(weekday_mapper))], weekday_mapper))
data['weekday'] = data['weekday'].map(weekday_mapper)
data['hour'] = data['event_time'].dt.hour
data = data.drop(columns=['event_time'])
# data
```

## 季度运营情况分析

### 季度运营指标汇总

#### 流量指标

指标口径：
- pv：浏览量，‘view’次数
- uv：独立访客数
- 跳失率：仅有 1 次浏览行为的用户比例
- 人均浏览次数：平均每个用户浏览次数

```python
pv = data['view'].sum()
uv = data['user_id'].nunique()
user_behaviour_gb = data[['user_id', 'view', 'cart', 'remove_from_cart', 'purchase']].groupby('user_id').sum()
bounce_rate = len(user_behaviour_gb[(user_behaviour_gb['view'] == 1) & \
                                    (user_behaviour_gb['cart'] + user_behaviour_gb['remove_from_cart'] + user_behaviour_gb['purchase'] == 0)]) / len(user_behaviour_gb)
per_capita_view = user_behaviour_gb['view'].mean()

headers = ['PV', 'UV', '跳失率', '人均浏览次数']
flow= [[pv, uv, f"{bounce_rate:.2%}", f"{per_capita_view:.2f}"]]
flow_table = (
    Table()
    .add(headers, flow)
)
flow_table.render_notebook()
```

#### 行为指标

指标口径：

- 加购次数：'cart'次数
- 移出购物车次数：'remove_from_cart'次数
- 购买次数：'purchase'次数
- 季度平均使用天数：每日登陆用户数（去重）之和与季度登陆用户数（去重）的比值

```python
cart_n, rmc_n, purchase_n = data['cart'].sum(), data['remove_from_cart'].sum(), data['purchase'].sum()
quarter_average_login = len(data[['user_id', 'date']].drop_duplicates()) / data['user_id'].nunique()

headers = ['加购次数', '移出购物车次数', '购买次数', '季度平均使用天数']
behaviour_sum = [[cart_n, rmc_n, purchase_n, f"{quarter_average_login:.2f}"]]
behaviour_sum_table = (
    Table()
    .add(headers, behaviour_sum)
)
behaviour_sum_table.render_notebook()
```

```python
convert_df = data[['view', 'cart', 'purchase', 'remove_from_cart']].sum()
convert_df = np.round(convert_df / convert_df[0] * 100, 2)

colors = ['#D53A35', '#2F4554', '#61A0A8', '#D48265', '#749F83', '#CA8622', '#BDA29A', '#6E7074']

behaviour_nodes = [
    {"name": "view"},
    {"name": "cart"},
    {"name": "remove_from_cart"},
    {"name": "purchase"},
    {"name": "miss"},
]

behaviour_links = [
    {"source": "view", "target": "cart", "value": convert_df['cart']},
    {"source": "view", "target": "miss", "value": np.round(convert_df['view'] - convert_df['cart'], 2)},
    {"source": "cart", "target": "purchase", "value": convert_df['purchase']},
    {"source": "cart", "target": "remove_from_cart", "value": convert_df['remove_from_cart']},
    {"source": "cart", "target": "miss", "value": np.round(convert_df['cart'] - convert_df['purchase'] - convert_df['remove_from_cart'], 2)},
]

behaviour_sankey = (
    Sankey()
    .set_colors(colors=colors)
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="整体行为转化流向",
            subtitle="注：转化率以【浏览量】为计算基准。",
        ),
        tooltip_opts=opts.TooltipOpts(trigger="item", trigger_on="mousemove"),
        
    )
    .add(
        "", 
        behaviour_nodes, 
        behaviour_links,
        orient="vertical",
        pos_top="10%",
        pos_left="10%",
        pos_right="10%",
        focus_node_adjacency="allEdges",
        linestyle_opt=opts.LineStyleOpts(opacity=0.4, curve=0.4, color="target"),
    )
)
behaviour_sankey.render_notebook()
```

结论：
- “浏览-加购”的转化率为 60.34%，反映出用户浏览后多半会使用加购功能。
- “加购-移出购物车”的整体转换率为 31.43%，是“加购-购买”的整体转化率 13.83% 的两倍不止，猜测用户加购的目的不单是计划购买，还可能是商品收藏或价格观察，所以加购后又予以移除。

建议：
- 推出收藏功能，既完善用户使用的功能性，又便于我们分析用户的意图。
- 适时调整商品价格，促进用户购买。例如：发放优惠券促销、清仓打折。
- 当加购后的商品发生信息变更或久未操作，及时提醒用户。例如：客户端-推送、web 端-消息通知。

#### 交易指标

指标口径：
- 销售额：购买物品的价格总数
- 客单价：销售额除以购买客户数
- 客户数：有购买记录的用户数
- 购买率：所有用户中，有购买记录的用户占比
- 复购客户数：有购买记录的用户中，有两次以上购买记录的用户
- 客户复购率：有购买记录的用户中，有两次以上购买记录的用户占比
- 平均购买次数：购买用户的平均购买次数
- 商品复购率：客户重复购买某件商品的比例

```python
turnovers = data['turnovers'].sum()
uppc = data[data['purchase'] == 1].groupby(by=['user_id', 'product_id'])[['purchase']].count()
ups = uppc.groupby('user_id').sum()
customer_unit_price = turnovers / ups.shape[0]
purchase_rate = len(ups) / uv
repurchase_n = len(ups[ups['purchase'] > 1])
repurchase_rate = repurchase_n / ups.shape[0]

headers = ['销售额', '客单价', '客户数', '用户购买率', '复购客户数', '客户复购率', '客户平均购买次数', '商品复购率',]
transaction = [[
    f"{turnovers:.2f}",
    f"{customer_unit_price:.2f}",
    f"{ups.shape[0]}",
    f"{purchase_rate:.2%}",
    f"{repurchase_n}",
    f"{repurchase_rate:.2%}",
    f"{ups['purchase'].mean():.0f}",
    f"{(uppc['purchase'] > 1).sum() / uppc.shape[0]:.2%}",
]]
transaction_table = (
    Table()
    .add(headers, transaction)
)
transaction_table.render_notebook()
```

**结论：复购率高达 89.20%，客户粘性很高。**

### 用户活跃情况

用户行为统计均值：

```python
def u_3sigma(data):
    data_outliars = pd.DataFrame()
    for f in data.columns:
        u, sigma = data[f].mean(), data[f].std()
        lower, upper = u - 3 * sigma, u + 3 * sigma
        data_outliars[f] = data[f].apply(lambda x: False if x >= lower and x <= upper else True)
    return data_outliars


user_gb = data[['user_id', 'view', 'cart', 'remove_from_cart', 'purchase', 'turnovers']].groupby('user_id').sum()
user_outliars = u_3sigma(user_gb)
user_clear = user_gb[~user_outliars.any(axis=1)]
# user_clear

user_bar = (
    Bar()
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='用户行为统计均值',
            subtitle='注：清洗用户 -- 行为数据 ∈ [u-3σ, u+3σ] 的用户。'
        )
    )
    .add_xaxis(user_gb.columns.tolist())
    .add_yaxis('整体用户', user_gb.describe().loc['mean'].tolist(), label_opts=opts.LabelOpts(is_show=False))
    .add_yaxis('清洗用户', user_clear.describe().loc['mean'].tolist(), label_opts=opts.LabelOpts(is_show=False))
    .add_yaxis('无购买用户', user_gb[user_gb['purchase'] == 0].describe().loc['mean'].tolist(), label_opts=opts.LabelOpts(is_show=False))
    .add_yaxis('有购买用户', user_gb[user_gb['purchase'] > 0].describe().loc['mean'].tolist(), label_opts=opts.LabelOpts(is_show=False))
)
user_bar.render_notebook()
```

现象：购买用户和未购买用户的各项行为指标存在明显差异。

建议：提升用户活跃度以促进购买转化。

结论：部分用户仅在活动期活跃。

建议：合理安排活动期，同时做好拉新与召回。

    2. 通过社交媒体、广告增大曝光量，吸引新用户

    2. 针对沉默用户（长周期内使用过，短周期内未使用）：通过内容推荐、定期活动推送召回

```python
user_new_day = data.groupby('user_id')[['date']].min()
user_new_day.columns = ['first_day']
user_day = pd.merge(user_new_day, data[['user_id', 'date']], on='user_id', how='left')
user_day['nday'] = user_day['date'] - user_day['first_day']
user_day = user_day[user_day['nday'].dt.days < 31]  # 只取 30 天留存
day_retention_n = user_day[['first_day', 'nday', 'user_id']].groupby(['first_day', 'nday'], as_index=False).count()

retention_table = day_retention_n.pivot_table(index=['first_day'], columns=['nday'])
retention_table.columns = [i for i in range(retention_table.shape[1])]
first_day_n = retention_table[0]
del retention_table[0]
for col in retention_table.columns:
    retention_table[col] = np.round(retention_table[col] / first_day_n, 4)

retention_table = retention_table.iloc[::-1].T
retention_values = [[i, j, val] for i, row in enumerate(retention_table.values) for j, val in enumerate(row)]
max_retention = np.max(retention_table)

retention_heatmap = (
    HeatMap()
    .set_global_opts(
        title_opts=opts.TitleOpts(title='每日用户留存（30日）'),
        xaxis_opts=opts.AxisOpts(
            name='第 N 日',
            position='top'
        ),
        yaxis_opts=opts.AxisOpts(
            name='日期',
            name_location='start',
            axislabel_opts=opts.LabelOpts(font_size=9),
        ),
        visualmap_opts=opts.VisualMapOpts(
            max_=np.round(np.max(retention_table.iloc[0]) + 0.1, 1),
            is_calculable=True,
            pos_top='middle',
            pos_left='right',
            precision=2,
        ),
    )
    .add_xaxis(retention_table.index.astype(str).tolist())
    .add_yaxis("第 N 日留存率", retention_table.columns.astype(str).tolist(), retention_values)
)
retention_heatmap.render_notebook()
```

现象：次日留存在 0.1~0.2 之间，第一日往后的留存基本都小于 0.1，留存率整体较低且变化不明显，在活动期间有微小提升。

结论：对于电商，大多数用户在有购物需求或活动期才会使用，用户粘性低于具有日常高频的使用需求的社交/游戏类产品，因而难有稳定且高的留存率。

```python
user_view_cnt = data[['user_id', 'view', 'turnovers']].groupby('user_id').sum()
user_login_days = data[['user_id', 'date']].drop_duplicates().groupby('user_id').count()
user_login_days.columns = ['login_days']
user_stickiness = pd.concat([user_view_cnt, user_login_days], axis=1)
user_stickiness['average_daily_view'] = user_stickiness['view'] / user_stickiness['login_days']
login_days_df = user_stickiness.groupby('login_days')[['turnovers']].mean()
login_days_df.index.name = '累计登陆天数'
login_days_df.columns = ['平均销售额']
login_days_df['用户比例'] = user_stickiness['login_days'].value_counts(normalize=True).sort_index()
login_days_df['平均日均访问量'] = user_stickiness.groupby('login_days')[['average_daily_view']].mean()
double_y(login_days_df, left=['用户比例'], right=['平均销售额', '平均日均访问量'], title='用户累计登陆天数分布')
```

**结论：用户日常粘性有待提高，以期提高单个用户的 LTV。**

**结论：排除均价的因素，品牌流量的高低几乎决定销量的高低。近 80% 的品牌处于低流量-低销量的状况，可从提高流量的角度入手改善。**

建议：
1. 尽可能增加品牌的曝光度并转化为有效流量，提高流量几乎是提高销量的必然前提。
2. 提高流量的同时，合理控制价格，以期销售额的最大化。
