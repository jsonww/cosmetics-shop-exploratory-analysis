## 季度整体情况

### 季度指标汇总

#### 流量指标

指标口径：
- pv：浏览量，‘view’次数
- uv：独立访客数
- 跳失率：仅有 1 次浏览行为的用户比例
- 人均浏览次数：平均每个用户浏览次数

```python
pv = data['view'].sum()
uv = data['user_id'].nunique()
user_behaviour_gb = data[['user_id', 'view', 'cart', 'remove_from_cart', 'purchase']].groupby('user_id').sum()
bounce_rate = len(user_behaviour_gb[(user_behaviour_gb['view'] == 1) & \
                                    (user_behaviour_gb['cart'] + user_behaviour_gb['remove_from_cart'] + user_behaviour_gb['purchase'] == 0)]) / len(user_behaviour_gb)
per_capita_view = user_behaviour_gb['view'].mean()

headers = ['PV', 'UV', '跳失率', '人均浏览次数']
flow= [[pv, uv, f"{bounce_rate:.2%}", f"{per_capita_view:.2f}"]]
flow_table = (
    Table()
    .add(headers, flow)
)
flow_table.render_notebook()
```

#### 行为指标

指标口径：

- 加购次数：'cart'次数
- 移出购物车次数：'remove_from_cart'次数
- 购买次数：'purchase'次数
- 平均使用天数：每日登陆用户数（去重）之和与季度登陆用户数（去重）的比值

```python
cart_n, rmc_n, purchase_n = data['cart'].sum(), data['remove_from_cart'].sum(), data['purchase'].sum()
month_average_login = len(data[['user_id', 'date']].drop_duplicates()) / data['user_id'].nunique()

headers = ['加购次数', '移出购物车次数', '购买次数', '平均使用天数']
behaviour_sum = [[cart_n, rmc_n, purchase_n, f"{month_average_login:.2f}"]]
behaviour_sum_table = (
    Table()
    .add(headers, behaviour_sum)
)
behaviour_sum_table.render_notebook()
```

#### 交易指标

指标口径：
- 成交额：购买物品的价格总数
- 客单价：成交额除以购买客户数
- 客户数：有购买记录的用户数
- 购买率：所有用户中，有购买记录的用户占比
- 复购客户数：有购买记录的用户中，有两次以上购买记录的用户
- 复购率：有购买记录的用户中，有两次以上购买记录的用户占比
- 平均购买次数：购买用户的平均购买次数

```python
turnovers = data['turnovers'].sum()
user_gb = data[['user_id', 'view', 'cart', 'remove_from_cart', 'purchase', 'turnovers']].groupby('user_id').sum()
buyer_gb = user_gb[user_gb['purchase'] > 0]
customer_unit_price = turnovers / buyer_gb.shape[0]
purchase_rate = len(buyer_gb) / len(user_gb)
repurchase_n = len(buyer_gb[buyer_gb['purchase'] > 1])
repurchase_rate = repurchase_n / len(buyer_gb)

headers = ['成交额', '客单价', '客户数', '购买率', '复购客户数', '复购率', '客户平均购买次数']
transaction = [[
    f"{turnovers:.2f}",
    f"{customer_unit_price:.2f}",
    f"{buyer_gb.shape[0]}",
    f"{purchase_rate:.2%}",
    f"{repurchase_n}",
    f"{repurchase_rate:.2%}",
    f"{buyer_gb['purchase'].mean():.0f}"
]]
transaction_table = (
    Table()
    .add(headers, transaction)
)
transaction_table.render_notebook()
```

**结论：整体购买率仅 7.39%，新用户的运营有改善空间；复购率高达 89.20%，老客户的粘性很高。**

### 整体行为转化

```python
def draw_behaviour_funnel(data, behaviour_lst, overall=True, is_show=True, title='Funnel'):
    behaviour_df = pd.DataFrame(data[behaviour_lst].sum(), index=behaviour_lst, columns=['n'])
    if overall:
        behaviour_df['转化率'] = behaviour_df['n'] / behaviour_df.iloc[0, 0]
    else:
        behaviour_df['转化率'] = behaviour_df['n'].shift(-1) / behaviour_df['n']
        behaviour_df['转化率'] = behaviour_df['转化率'].shift(1)
        behaviour_df.iloc[0, 1] = 1
    behaviour_df['转化率'] = behaviour_df['转化率'].apply(lambda x: format(x, ".2%"))
    behaviour_df['转化环节'] = behaviour_df.index + ': ' + behaviour_df['转化率']

    behaviour_funnel = (
        Funnel()
        .add(
            '',
            behaviour_df[['转化环节', 'n']].values,
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=title,
                subtitle='转化率计算方式：每个环节以第一个环节作为计算基准' if overall else '计算方式：每个环节以前一个环节作为计算基准',
            ),
            legend_opts=opts.LegendOpts(
                pos_left='right',
                pos_top='bottom',
                orient='vertical'
            ),
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(
                position='inner',
                formatter='{b}',
            )
        )
    )
    return behaviour_funnel.render_notebook() if is_show else behaviour_funnel
```

```python
def show_time_trend(uv_df, behaviour_df, bar_title='UV', bar_subtitle='-', line_title='Behaviour', line_subtitle='-', datazoom_opts_show=False, is_show=True):
    """Show the data trend of specific timescales."""
    uv_bar = (
        Bar()
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=bar_title,
                subtitle=bar_subtitle,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
        )
        .add_xaxis(uv_df.index.astype('str').tolist())
    )
    for c in uv_df.columns:
        uv_bar.add_yaxis(c, uv_df[c].tolist())
        uv_bar.set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            ),
            markline_opts=opts.MarkLineOpts(
                data=[
                    opts.MarkLineItem(name=f"{c}均值", type_="average")
                ],
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
                linestyle_opts=opts.LineStyleOpts(
                    opacity=0.5,
                    type_='dashed'
                ),
            ),
        )
    if datazoom_opts_show:
        uv_bar.set_global_opts(datazoom_opts=opts.DataZoomOpts(is_show=datazoom_opts_show))
    
    behaviour_line = (
        Line()
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=line_title,
                subtitle=line_subtitle,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
        )
        .add_xaxis(behaviour_df.index.astype(str).tolist())
    )
    for b in behaviour_df.columns:
        behaviour_line.add_yaxis(b, behaviour_df[b].tolist())
        behaviour_line.set_series_opts(
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=2),
            markpoint_opts=opts.MarkPointOpts(
                data=[
                    opts.MarkPointItem(type_="max", name="最大值"),
                    opts.MarkPointItem(type_="min", name="最小值"),
                ]
            ),
            markline_opts=opts.MarkLineOpts(
                data=[
                    opts.MarkLineItem(name=f"{b}均值", type_="average")
                ],
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
                linestyle_opts=opts.LineStyleOpts(
                    opacity=0.5,
                    type_='dashed'
                ),
            ),
        )

    page = (
        Page()
            .add(uv_bar, behaviour_line)
    )
    return page.render_notebook() if is_show else page
```

```python
def double_y(data, left, right, title='Double y-axis', left_kind=Line(), right_kind=Line(), is_show=True):
    """双 y 轴"""
#     kind_mapper = {
#         'Line': Line(),
#         'Bar': Bar(),
#     }
#     left_chart = kind_mapper[left_kind] 
#     right_chart = kind_mapper[right_kind]
    
    left_chart = (
        Line()
        .add_xaxis(data.index.astype('str').tolist())
        .set_global_opts(
            title_opts=opts.TitleOpts(
                title=title,
            ),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                name=f"{data.index.name}",
                name_location='middle',
                name_gap=30,
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
            yaxis_opts=opts.AxisOpts(
                name=f"{'/'.join(left)}",
            ),
        )
        .extend_axis(
            yaxis=opts.AxisOpts(
                name=f"{'/'.join(right)}",
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            )
        )
    )

    for l in left:
        left_chart.add_yaxis(
            l,
            data[l].tolist(),
            label_opts=opts.LabelOpts(is_show=False),
            linestyle_opts=opts.LineStyleOpts(width=2),
        )

    right_chart = (
        Line()
        .add_xaxis(data.index.astype('str').tolist())
    )

    for r in right:
        right_chart.add_yaxis(
            r,
            data[r].tolist(),
            yaxis_index=1,
            linestyle_opts=opts.LineStyleOpts(curve=360, type_='dotted'),
            label_opts=opts.LabelOpts(is_show=False),
        )

    left_chart.overlap(right_chart)
    return left_chart.render_notebook() if is_show else left_chart
```

```python
def u_3sigma(data):
    data_outliars = pd.DataFrame()
    for f in data.columns:
        u, sigma = data[f].mean(), data[f].std()
        lower, upper = u - 3 * sigma, u + 3 * sigma
        data_outliars[f] = data[f].apply(lambda x: False if x >= lower and x <= upper else True)
    return data_outliars


user_outliars = u_3sigma(user_gb)
user_clear = user_gb[~user_outliars.any(axis=1)]
user_clear
user_clear.describe()
```

```python
def draw_pareto(data, tittle='帕累托图', x_name='', is_show=True):
    '''
    Draw pareto.
    '''
    data[f'{data.columns[0]}-累计占比'] = np.round(data.values[:, 0].cumsum() / data.values[:, 0].sum(), 2)
    
    bar = (
        Bar()
        .set_global_opts(
            title_opts=opts.TitleOpts(title=tittle),
            tooltip_opts=opts.TooltipOpts(
                is_show=True, trigger="axis", axis_pointer_type="cross"
            ),
            xaxis_opts=opts.AxisOpts(
                name=x_name,
                name_location='center',
                name_gap=30,
                type_="category",
                axispointer_opts=opts.AxisPointerOpts(is_show=True, type_="shadow"),
            ),
            yaxis_opts=opts.AxisOpts(
                name=data.columns[0],
                type_="value",
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
                axistick_opts=opts.AxisTickOpts(is_show=True),
            ),
            legend_opts=opts.LegendOpts(
                type_="scroll",
            ),
        )
        .extend_axis(
            yaxis=opts.AxisOpts(
                name=data.columns[1],
                type_="value",
                min_=0,
                max_=1.1,
                axislabel_opts=opts.LabelOpts(formatter="{value}"),
            )
        )
        .add_xaxis(data.index.astype(str).tolist())
        .add_yaxis(
            data.columns[0], 
            data.iloc[:, 0].tolist(),
            z=0
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )

    cumsum_line = (
        Line()
        .add_xaxis(data.index.astype(str).tolist())
        .add_yaxis(
            data.columns[1], 
            data.iloc[:, 1].tolist(),
            yaxis_index=1,
            linestyle_opts=opts.LineStyleOpts(width=1),
            markline_opts=opts.MarkLineOpts(
                data=[opts.MarkLineItem(y=0.8)],
                linestyle_opts=opts.LineStyleOpts(
                    width=1,
                    type_='dashed',
                    opacity=0.8,
                ),
                symbol='none',
                label_opts=opts.LabelOpts(is_show=False),
            ),
            z=1
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    )
    
    return bar.overlap(cumsum_line).render_notebook() if is_show else bar.overlap(cumsum_line)
```

### 客户分类

#### 传统 RFM 模型

构建 RFM 模型，对客户（本季度有过消费行为的用户）进行分层，方便跟踪各层用户的特征及变化情况，以针对性运营提高效率。

目标指标：
- Recency: 最近一次消费时间间隔
- Frequency：消费频率
- Monetary：消费金额

划分方式：以均值作为划分标准，将 RFM 三个指标数据转化为高、低两个等级变量，根据用户每项指标等级，划分为 8 类具不同消费特征的群体，由此直观反映各类客户的价值和重要性，并可根据相应特征采取针对性策略，可操作性强。

|R|F|M|划分群体类型|特点|
|-|-|-|-|-|
|高|高|高|重要价值客户|高端客户，各项指标都优于平均，质量最优|
|低|高|高|重要保持客户|最近购买日期早于平均最近购买日期，存在流失的可能性|
|高|低|高|重要发展客户|消费次数低于均值，用户粘性不高|
|低|低|高|重要挽留客户|消费次数低于均值和最近购买日期早于平均最近购买日期，流失风险大|
|高|高|低|一般价值客户|近期消费、频率高于平均，但消费金额低于平均|
|低|高|低|一般保持客户|频率高于平均、消费金额低于平均，近期无消费，存在流失的可能性|
|高|低|低|一般发展客户|近期消费，频率和消费金额均低于平均，可能是试水新用户或粘性不高的老用户|
|低|低|低|一般挽留客户|各项指标都低于平均|

注：由于消费金额 M 是衡量用户价值更重要的属性，根据消费高金额 M 的高低又划分为重要客户和一般客户。

```python
# 统计消费用户 rfm 指标
user_rfm = pd.DataFrame(data['user_id'].unique(), columns=['user_id'])
user_recently = pd.DataFrame(data[data['purchase'] == 1].groupby('user_id', as_index=False)['date'].apply(lambda x: x.max()))
first_date = data['date'].min()
user_recently['date'] = (user_recently['date'] - first_date).dt.days
user_rfm = pd.merge(left=user_rfm, right=user_recently, how='left', on=['user_id'])
user_frequency = data[['user_id', 'purchase']].groupby('user_id').sum()
user_rfm = pd.merge(left=user_rfm, right=user_frequency, how='left', on=['user_id'])
user_monetary = data[['user_id', 'turnovers']].groupby('user_id').sum()
user_rfm = pd.merge(left=user_rfm, right=user_monetary, how='left', on=['user_id'])
user_rfm.fillna(0, inplace=True)
user_rfm.set_index('user_id', inplace=True)
user_rfm.columns = ['Recently','Frequency', 'Monetary']
# user_rfm
```

```python
consumer_rfm = user_rfm[user_rfm['Frequency'] > 0]

# 划分 rfm 等级变量
rfm_divide = {col: consumer_rfm[col].mean() for col in consumer_rfm.columns}  # 取 rfm 中位数
consumer_rfm = pd.DataFrame(consumer_rfm)

for col in consumer_rfm:
    '''等级变量'''
    consumer_rfm[f"_{col}"] = consumer_rfm[col].apply(lambda x: 1 if x > rfm_divide[col] else 0)

# 根据 rfm 等级变量打标签
consumer_rfm['code'] = 100 * consumer_rfm['_Recently'] + 10 * consumer_rfm['_Frequency'] + consumer_rfm['_Monetary']  # 编码
rfm_code2label = {
    111: '重要价值客户',
    11: '重要保持客户',
    101: '重要发展客户',
    1: '重要挽留客户',
    110: '一般价值客户',
    10: '一般保持客户',
    100: '一般发展客户',
    0: '一般挽留客户',
}
consumer_rfm['label'] = consumer_rfm['code'].apply(lambda x: rfm_code2label[x])  # 编码转标签

# 3. 分层效果检验
# 3.1 各类用户占比
labels_order = ['重要价值客户', '重要保持客户', '重要发展客户', '重要挽留客户', '一般价值客户', '一般保持客户', '一般发展客户', '一般挽留客户']
label_vc = consumer_rfm['label'].value_counts().reindex(labels_order)

rfm_pie = (
    Pie()
    .add(
        '',
        [list(z) for z in zip(label_vc.index.tolist(), label_vc.values.tolist())],
        radius=[80, 180],
        rosetype="radius", #玫瑰饼图
    )
    .set_global_opts(
        title_opts=opts.TitleOpts('RFM 分层效果'),
        legend_opts=opts.LegendOpts(is_show=False)
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {d}%'))
)

# 3.2 各类用户 rfm 雷达图
label_mean = consumer_rfm.groupby('label')[['Recently', 'Frequency', 'Monetary']].mean()
label_mean = label_mean.reindex(labels_order)
colors = ['#D53A35', '#2F4554', '#61A0A8', '#D48265', '#749F83', '#CA8622', '#BDA29A', '#6E7074']
# 数据做归一化处理
label_mean_mms = MinMaxScaler().fit_transform(label_mean)

label_mean_radar = (
    Radar()
    .add_schema(
        schema=[opts.RadarIndicatorItem(name=col) for col in label_mean.columns]
    )
    .set_global_opts(
        title_opts=opts.TitleOpts("各类客户 RFM 雷达图"),
        legend_opts=opts.LegendOpts(pos_bottom='15%'),
    )
)

for i, v in enumerate(label_mean.index):
    label_mean_radar.add(v, [label_mean.loc[v].tolist()], color=colors[i])
    label_mean_radar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    
label_monetary = pd.DataFrame(consumer_rfm.groupby('label')['Monetary'].sum()).reindex(labels_order)
label_monetary.columns = ['消费金额']
pareto = draw_pareto(label_monetary, x_name='客户类型', is_show=False)
    
rfm_page = (
    Page()
    .add(rfm_pie, label_mean_radar, pareto)
)
rfm_page.render_notebook()
```

- 一般客户占比约 70%，贡献整体销售额的 34%。其中，一般发展客户（32.51%）和一般挽留客户（29.51%）在整体客户中占比明显高于其它群体，是最大的两个客户群体。

**结论：客户质量两极分化，一般客户基数较大有待开发。**

建议：
- 经营好重要客户，保持重要客户的粘性。例如：
    - '重要价值客户'： VIP 优待 -- 提供更优质的售时售后服务，或进行不定期福利回馈。
    - '重要保持客户'：及时唤醒 -- 采取措施重新唤醒，促进客户复购。若存有联系方式，而可通过固定周期未购买进行消息通知。
    - '重要发展客户' 和 '重要挽留客户'：提高客户粘性 -- 增加对这类客户的曝光量和吸引力。前者可以及时推送优惠活动，后者可以通过调研客户消费需求和满意度调查，对能够改进的地方进行完善。
- 充分开发一般客户，使其向更高层客户群体次转化。可以针对这部分客户的进行消费需求调研和满意度调查，对能够改进的地方进行完善。

#### KMeans 聚类

```python
consumer_kmeans = user_rfm[user_rfm['Frequency'] > 0]
consumer_ss = MinMaxScaler().fit_transform(consumer_kmeans)

# Calinski-Harabasz Index 评估自动取最高值的 K 为类数
best_k, best_calinski = 0, 0
for k in range(2, 9):
    y_pred = KMeans(n_clusters=k, random_state=1234).fit_predict(consumer_ss)
    calinski = metrics.calinski_harabasz_score(consumer_ss, y_pred)
#     print(f"{k}: {calinski}.")
    if calinski > best_calinski:
        best_k, best_calinski = k, calinski

# 使用自选择最佳 K 进行聚类
best_model = KMeans(n_clusters=best_k, random_state=1234)
best_model.fit(consumer_ss)
consumer_kmeans['label'] = best_model.labels_
label_mean = consumer_kmeans.groupby('label').mean().sort_values(by=['Monetary', 'Frequency', 'Recently'], ascending=False)
label_mapper = dict(zip(label_mean.index, [chr(ord('A') + i) for i in range(label_mean.shape[0])]))
consumer_kmeans['label'] = consumer_kmeans['label'].map(label_mapper)
label_mean.index = [chr(ord('A') + i) for i in range(label_mean.shape[0])]
label_vc = consumer_kmeans['label'].value_counts().reindex(labels_order)
colors = ['#D53A35', '#2F4554', '#61A0A8', '#D48265', '#749F83', '#CA8622', '#CA8622', '#6E7074']

kmeans_pie = (
    Pie()
        .add(
        '',
        [list(z) for z in zip(label_vc.index.tolist(), label_vc.values.tolist())],
        radius=[100, 180]
    )
        .set_global_opts(
        title_opts=opts.TitleOpts('K-Means 分层效果'),
        legend_opts=opts.LegendOpts(is_show=False)
    )
        .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}: {d}%'))
)


label_mean_mms = MinMaxScaler().fit_transform(label_mean)
label_mean_radar = (
    Radar()
    .add_schema(
        schema=[opts.RadarIndicatorItem(name=col) for col in label_mean.columns]
    )
    .set_global_opts(
        title_opts=opts.TitleOpts("各类用户雷达图"),
        legend_opts=opts.LegendOpts(
        pos_right='0%',
            pos_top='middle',
            orient='vertical',
        ),
    )
)

for i, v in enumerate(label_mean.index):
    label_mean_radar.add(v, [label_mean.loc[v].tolist()], color=colors[i])
    label_mean_radar.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    
label_monetary = pd.DataFrame(consumer_kmeans.groupby('label')['Monetary'].sum())
label_monetary.columns = ['贡献销售额']
pareto = draw_pareto(label_monetary, is_show=False)

kmeans_page = (
    Page()
    .add(kmeans_pie, label_mean_radar, pareto)
)
kmeans_page.render_notebook()
```

现象：使用 KMeans 基于 Calinski-Harbasz Score 自动选择最佳聚类数为 6 类。分层属性中，各类用户“最近购买时间”的差异较大，而“购买金额”和“购买频次”差异较小。

**结论：对“最近购买时间”较远的用户群体应及时进行召回。**
